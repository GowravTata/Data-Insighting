```python
#most number of 6s in an innings
n=a[a['Batsman_Scored']==6]
nu=n.groupby(['Striker_Id','Match_Id'])['Batsman_Scored'].count().reset_index().sort_values(by='Batsman_Scored',ascending=False)
num=pd.merge(nu,c,left_on='Striker_Id',right_on='Player_Id',how='inner')
num
```

```python
#most number of 4s in an innings
f=a[a['Batsman_Scored']==4]
fu=f.groupby(['Striker_Id','Match_Id'])['Batsman_Scored'].count().reset_index().sort_values(by='Batsman_Scored',ascending=False)
fur=pd.merge(fu,c,left_on='Striker_Id',right_on='Player_Id',how='inner')
fur
```

```python
#Which player scored the most number of centuries
m=a.groupby(['Striker_Id','Match_Id'])['Batsman_Scored'].sum().reset_index()
mm=m[m['Batsman_Scored']>=100]
mm1=mm.groupby(['Striker_Id'])['Batsman_Scored'].count()
mm2=pd.merge(mm1,c,left_on='Striker_Id',right_on='Player_Id')
mm2=mm2.sort_values(by='Batsman_Scored',ascending=False)
mm2
```

```python
#total number of centuries
cen=a.groupby(['Striker_Id','Match_Id'])['Batsman_Scored'].sum().reset_index()
cent=cen[cen['Batsman_Scored']>=100].count()
cent
```

```python
#total number of fifties
fif=a.groupby(['Striker_Id','Match_Id'])['Batsman_Scored'].sum().reset_index()
fift=fif[fif['Batsman_Scored']>=50].count()
fift
```

```python
#Which player scored the most number of fifties
n=a.groupby(['Striker_Id','Match_Id'])['Batsman_Scored'].sum().reset_index()
nn=n[n['Batsman_Scored']>=50]
nn1=nn.groupby(['Striker_Id'])['Batsman_Scored'].count()
nn2=pd.merge(nn1,c,left_on='Striker_Id',right_on='Player_Id')
nn2=nn2.sort_values(by='Batsman_Scored',ascending=False)
nn2
```

```python
##highest run-scorer in all the seasons
s=a.groupby(['Striker_Id'])['Batsman_Scored'].sum()
s=s.reset_index().sort_values(by='Batsman_Scored',ascending=False)
f1=pd.merge(s,c,left_on='Striker_Id',right_on='Player_Id',how='inner')
f1.head()
```

```python
#Find the 3rd most running scored player from overseas
f2=f1[f1['Country']!='India']
f2=f2[['Player_Name','Batsman_Scored']]
f2.head()
```

```python
#most dot balls bowled by a bowler
aa=a[a['Batsman_Scored']==0]
aa=aa.groupby(['Bowler_Id'])['Batsman_Scored'].count().reset_index()
dot=pd.merge(aa,c,left_on='Bowler_Id',right_on="Player_Id",how='inner')
dot1=dot[['Player_Name','Batsman_Scored']].sort_values(by='Batsman_Scored',ascending=False)
dot1.rename(columns={'Batsman_Scored':'Total_Dot_Balls'},inplace=True)
dot1.head()
```
